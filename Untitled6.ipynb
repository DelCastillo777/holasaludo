Introduccion teorica:
Con el conjunto de datos que se entrega realizar practica de regresion lineal simple y multiple.

Variable independientes (ej. peso, caballos de fuerza).
Variable depedientes (ej. consumo: mpg).
Carga de datos: Se optiene desde un CSV en linea (dataset Auto MPG).
Exploracion inicial: Mostrar primeras filas, info y estadisticas.
Regresion Lineal Simple: Ejemplo:mpg -weight + horsepower + displacement.Evaluacion con R al cuadrado y RMSE.




---



```python
# Práctica completa: Regresión lineal simple y múltiple con gráficos usando Auto MPG
# Listo para ejecutar en Google Colab

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# 1. Carga de datos desde CSV local (sube tu archivo a Colab)
df = pd.read_csv('auto-mpg (2).csv')

# 2. Limpieza de datos: convertir '?' en NaN y asegurar que las columnas sean numéricas
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['displacement'] = pd.to_numeric(df['displacement'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['mpg'] = pd.to_numeric(df['mpg'], errors='coerce')

# 3. Exploración inicial
print("Primeras filas:")
print(df.head())
print("\nInformación general:")
print(df.info())
print("\nEstadísticas descriptivas:")
print(df.describe())

# 4. Regresión lineal simple (mpg vs weight)
X_simple = df[['weight']].dropna()
y_simple = df['mpg'].loc[X_simple.index]

model_simple = LinearRegression()
model_simple.fit(X_simple, y_simple)
y_pred_simple = model_simple.predict(X_simple)

r2_simple = r2_score(y_simple, y_pred_simple)
rmse_simple = np.sqrt(mean_squared_error(y_simple, y_pred_simple))

print("\nRegresión lineal simple (mpg vs weight):")
print(f"R²: {r2_simple:.2f}")
print(f"RMSE: {rmse_simple:.2f}")

# Gráfico de regresión lineal simple
plt.figure(figsize=(8,5))
plt.scatter(X_simple['weight'], y_simple, color='blue', alpha=0.5, label='Datos reales')
plt.plot(X_simple['weight'], y_pred_simple, color='red', label='Regresión lineal')
plt.title('Regresión lineal simple: mpg vs weight')
plt.xlabel('Peso (weight)')
plt.ylabel('Consumo (mpg)')
plt.legend()
plt.show()

# 5. Regresión lineal múltiple (mpg vs weight, horsepower, displacement)
df_multi = df[['weight', 'horsepower', 'displacement', 'mpg']].dropna()
X_multi = df_multi[['weight', 'horsepower', 'displacement']]
y_multi = df_multi['mpg']

model_multi = LinearRegression()
model_multi.fit(X_multi, y_multi)
y_pred_multi = model_multi.predict(X_multi)

r2_multi = r2_score(y_multi, y_pred_multi)
rmse_multi = np.sqrt(mean_squared_error(y_multi, y_pred_multi))

print("\nRegresión lineal múltiple (mpg vs weight, horsepower, displacement):")
print(f"R² múltiple: {r2_multi:.2f}")
print(f"RMSE múltiple: {rmse_multi:.2f}")

# Gráfico de comparación: valores reales vs predichos (múltiple)
plt.figure(figsize=(8,5))
plt.scatter(y_multi, y_pred_multi, color='green', alpha=0.5)
plt.plot([y_multi.min(), y_multi.max()], [y_multi.min(), y_multi.max()], 'r--', label='Predicción ideal')
plt.title('Regresión múltiple: mpg real vs mpg predicho')
plt.xlabel('mpg real')
plt.ylabel('mpg predicho')
plt.legend()
plt.show()
```
