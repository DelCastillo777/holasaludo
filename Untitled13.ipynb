```python
# --- CELDA 1: Importaciones y configuración de entorno ---
# Importa todas las librerías necesarias y configura la carpeta de salida.
import os
from datetime import datetime

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, Model, Input

print("TensorFlow:", tf.__version__)
print("GPUs detectadas:", tf.config.list_physical_devices("GPU"))
OUT = "/content/minimal_sintetico_unica_import"
os.makedirs(OUT, exist_ok=True)
print("Salida:", OUT)
print("Fecha (UTC):", datetime.utcnow().strftime("%Y-%m-%d %H:%M:%S"))
```

```python
# --- CELDA 2: Parámetros y semilla ---
# Define hiperparámetros y fija la semilla para reproducibilidad.
SEED = 123
np.random.seed(SEED)
tf.random.set_seed(SEED)

N_PER_CLASS = 300     # puntos por clase
NUM_CLASSES = 3       # número de clases (clusters)
TEST_FRAC = 0.2       # fracción para test
HIDDEN = 64           # neuronas en la capa oculta
EPOCHS = 40           # número de épocas
BATCH = 64            # tamaño de batch
VALIDATION_SPLIT = 0.15
```

```python
# --- CELDA 3: Generación y normalización de datos sintéticos ---
# Genera 3 clusters en 2D, mezcla, divide en train/test y normaliza.
centers = np.array([[-2.0, 0.0], [2.0, 0.0], [0.0, 2.5]])
X_list, y_list = [], []
for i, c in enumerate(centers):
    X_list.append(c + 0.8 * np.random.randn(N_PER_CLASS, 2))
    y_list.append(np.full(N_PER_CLASS, i, dtype=int))
X = np.vstack(X_list)
y = np.concatenate(y_list)
perm = np.random.permutation(len(X))
X, y = X[perm], y[perm]

split = int(len(X) * (1 - TEST_FRAC))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Guardar copia de test sin normalizar para la visualización de la frontera:
X_test_orig = X_test.copy()

# Normalizar usando media y std de train
mu, s = X_train.mean(0, keepdims=True), X_train.std(0, keepdims=True) + 1e-8
X_train = (X_train - mu) / s
X_test = (X_test - mu) / s

# Visualiza los datos sintéticos originales
plt.figure(figsize=(5, 4))
plt.scatter(X[:, 0], X[:, 1], c=y, s=10, cmap="tab10", edgecolor="k", linewidth=0.2, alpha=0.8)
plt.title("Datos sintéticos (clusters gaussianos)")
plt.xlabel("X1"); plt.ylabel("X2")
plt.grid(alpha=0.3)
plt.savefig(os.path.join(OUT, "datos.png"), dpi=120, bbox_inches="tight")
plt.show()
```

```python
# --- CELDA 4: Definición del modelo (MLP simple) ---
# Un MLP con una capa oculta, BatchNorm, Dropout y salida softmax.
inp = Input(shape=(2,), name="entrada_2d")
x = layers.Dense(HIDDEN, activation="relu", name="densa_1")(inp)
x = layers.BatchNormalization(name="bn_1")(x)
x = layers.Dropout(0.2, name="dropout_1")(x)
out = layers.Dense(NUM_CLASSES, activation="softmax", name="salida")(x)

model = Model(inputs=inp, outputs=out, name="mlp_minimal_unica_import")
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

model.summary()
```

```python
# --- CELDA 5: Entrenamiento y guardado del modelo ---
# Entrena usando EarlyStopping y guarda el modelo en formato .keras.
cb = [tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=6, restore_best_weights=True, verbose=1)]
hist = model.fit(
    X_train, y_train,
    validation_split=VALIDATION_SPLIT,
    epochs=EPOCHS,
    batch_size=BATCH,
    callbacks=cb,
    verbose=2
)

# Guardar en formato Keras nativo (.keras)
model.save(os.path.join(OUT, "modelo_minimal.keras"))

# Graficas de accuracy y loss
plt.figure(figsize=(8, 3))
plt.subplot(1, 2, 1)
plt.plot(hist.history.get("accuracy", []), label="train")
plt.plot(hist.history.get("val_accuracy", []), label="val")
plt.xlabel("Época"); plt.ylabel("Accuracy")
plt.legend(); plt.grid(True); plt.title("Accuracy")

plt.subplot(1, 2, 2)
plt.plot(hist.history.get("loss", []), label="train")
plt.plot(hist.history.get("val_loss", []), label="val")
plt.xlabel("Época"); plt.ylabel("Loss")
plt.legend(); plt.grid(True); plt.title("Loss")

plt.tight_layout()
plt.savefig(os.path.join(OUT, "curvas.png"), dpi=120, bbox_inches="tight")
plt.show()
```

```python
# --- CELDA 6: Evaluación y frontera de decisión ---
# Evalúa el modelo y grafica la frontera de decisión en el espacio original.
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Test loss: {loss:.4f}  |  Test accuracy: {acc:.4f}")

# Frontera de decisión: crea una malla en el espacio original
xx_min, xx_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
yy_min, yy_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.linspace(xx_min, xx_max, 300), np.linspace(yy_min, yy_max, 300))
grid = np.c_[xx.ravel(), yy.ravel()]
grid_norm = (grid - mu) / s
pred = model.predict(grid_norm, verbose=0)
lab = pred.argmax(axis=1).reshape(xx.shape)

# Dibuja los puntos de test en coordenadas originales (des-normalizados)
plt.figure(figsize=(6, 5))
plt.contourf(xx, yy, lab, alpha=0.3, cmap="coolwarm")
plt.scatter(X_test_orig[:, 0], X_test_orig[:, 1], c=y_test, s=18, cmap="tab10", edgecolor="k")
plt.title("Frontera de decisión y puntos de test")
plt.xlabel("X1"); plt.ylabel("X2")
plt.savefig(os.path.join(OUT, "frontera.png"), dpi=120, bbox_inches="tight")
plt.show()
```
