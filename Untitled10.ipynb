```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
#Cargamos el dataset
df = pd.read_csv('employee_data.csv')
df.head()
```

```python
#Seleccionamos variables relevantes
features = ['YearsExperience', 'MonthlySalary', 'PerformanceScore']
X = df[features]
X.head()
```

```python
#Normalizacion de datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Metodo del codo para encontrar el numero optimo de clusters
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, 'bo-')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo')
plt.show()
```

```python
#Elejir K optimo y crear los clusters
k_opt = 4 # Cambia este número según el resultado del codo
kmeans = KMeans(n_clusters=k_opt, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
#Vizualizaciones en 2D
sns.scatterplot(data=df, x='YearsExperience', y='MonthlySalary', hue='Cluster', palette='tab10')
plt.title('Clusters: Experiencia vs Salario')
plt.show()

sns.scatterplot(data=df, x='MonthlySalary', y='PerformanceScore', hue='Cluster', palette='tab10')
plt.title('Clusters: Salario vs Desempeño')
plt.show()
```

```python
#Vizualizacion en 3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    df['YearsExperience'], df['MonthlySalary'], df['PerformanceScore'],
    c=df['Cluster'], cmap='tab10', s=60
)
ax.set_xlabel('Años de Experiencia')
ax.set_ylabel('Salario Mensual')
ax.set_zlabel('Desempeño')
plt.title('Clusters 3D')
plt.show()
```

```python
#Interpretar los clusters
# Centroides normalizados
print(pd.DataFrame(kmeans.cluster_centers_, columns=features))

#Analizar caracteristicas de cada grupo
# Estadísticas de cada cluster
for c in range(k_opt):
    print(f"\nCluster {c}:")
    print(df[df['Cluster']==c][features].describe())
```
