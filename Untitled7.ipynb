```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creamos un dataset
data = {
    'edad': [22, 25, 47, 52, 46],
    'ingreso': [25000, 32000, 54000, 72000, 50000],
    'compra': [0, 1, 1, 0, 1]
}
df = pd.DataFrame(data)
df
```

```python
print(df.info())
print(df.describe())
```

```python
# Normalización simple
df['edad_norm'] = (df['edad'] - df['edad'].min()) / (df['edad'].max() - df['edad'].min())
df
```

```python
from sklearn.linear_model import LinearRegression

X = df[['edad']]  # Usamos solo 'edad' como ejemplo
y = df['ingreso']

modelo = LinearRegression()
modelo.fit(X, y)
print("Pendiente (coef):", modelo.coef_)
print("Intercepto:", modelo.intercept_)
```

```python
# Vector y matriz
v = np.array([1, 2])
M = np.array([[1, 2], [3, 4]])

# Producto punto
print("Producto punto:", np.dot(v, v))

# Norma de un vector
print("Norma:", np.linalg.norm(v))
```

```python
print("Media ingreso:", np.mean(df['ingreso']))
print("Mediana ingreso:", np.median(df['ingreso']))
```

```python
print("Varianza ingreso:", np.var(df['ingreso']))
print("Desviación estándar ingreso:", np.std(df['ingreso']))
```

```python
plt.hist(df['ingreso'], bins=5)
plt.title('Histograma de ingreso')
plt.xlabel('Ingreso')
plt.ylabel('Frecuencia')
plt.show()
```
