# APLICACION PRACTICA DE ALGEBRA LINEAL


```python
# Importar librerias
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt
```

```python
# cargar dataset
caras = fetch_olivetti_faces()
X = caras.data #imagenes
y = caras.target #etiquetas
y
# Mostrar un ejemplo de las imagenes
plt.imshow(caras.images[56], cmap='gray')
plt.title('ejemplo de rostro - Persona {y [0]}')
plt.show()
```

```python
# Aplicar analisis de componentes principales PCA
pca = PCA(n_components=100)
x_reducido = pca.fit_transform(X)
x_reconstruido = pca.inverse_transform(x_reducido)
```

```python
# Realizar la comporacion

# Seccionar una imagen de ejemplo
indice = 0
img_original = caras.images[indice]
img_reconstruida = x_reconstruido[indice].reshape(64, 64)

# Mostrar comparacion
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

axes[0].imshow(img_original, cmap='gray')
axes[1].imsshow(img_reconstruida, cmap='gray')
```
